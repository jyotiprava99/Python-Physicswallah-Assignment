**Q3. Program to print prime numbers between 0 to 50:**
```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print("Prime numbers between 0 and 50:")
for num in range(51):
    if is_prime(num):
        print(num, end=" ")
```

**Q9. Program to check if a number is an Armstrong number:**
```python
number = 371
num_str = str(number)
num_digits = len(num_str)
armstrong_sum = sum(int(digit)**num_digits for digit in num_str)
if armstrong_sum == number:
    print(number, "is an Armstrong number")
else:
    print(number, "is not an Armstrong number")
```

**Q10. Program to check if a year is a leap year:**
```python
year = 1996
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

**Q20. Find the difference between consecutive numbers in a list:**
```python
# Original list
numbers = [10, 20, 30, 40, 50]

# Calculate differences
differences = [second - first for first, second in zip(numbers[:-1], numbers[1:])]
print("Differences between consecutive numbers:", differences)
```

**Q23. Count the frequency of words in a text using a function:**
```python
def word_count(text):
    words = text.split()
    frequency = {}
    for word in words:
        frequency[word] = frequency.get(word, 0) + 1
    return frequency

text = "hello world hello"
print("Word count:", word_count(text))
```
