Q2: What is a scatter plot? Use the following code to generate data x and y. Using this
generated data plot a scatter plot

```python
import numpy as np
```

```python
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
plt.scatter(x,y)
plt.show()
```

A Scatter plot is a type of plot in which the values of two variables are plotted as points on a graph.
Each poit represents an observation in the data se, and the position of the point on the 
graph corresponds to the value of the two variables.

Q4 What is a Bar plot? why is it used? using the following data plot a bar plot and a horizontal bar plot.

Ans: Bar plot is a type of plot that displays categorical data using rectangular bars . Each bar represented a category and the height of the bar corresponds to the value or frequency of the category. 

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.title('Bar plot')
plt.show()
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.title("Horizontal bar plot")
plt.show()
```

Q5: What is box plot? Why it is used? Using the following data plot a box plot.

A Box plot also known as a box-and-whisker plot is atype of plot that displays the distribution of a dataset. It shows the median, quartiles and any outliers in the data.
Box plot is useful for vizualizing the spread and skewness and outliers in a dataset, making it easier to understand the overall distribution and identify any unusual observations.They are commomnly used in statistical analysis and data exploration.

```python
import seaborn as sns
box1 = np.random.normal(100, 10, 200)
#box2 = np.random.normal(90, 20, 200)
sns.boxplot(box1)
plt.title('Box plot 1')
plt.show()
```
